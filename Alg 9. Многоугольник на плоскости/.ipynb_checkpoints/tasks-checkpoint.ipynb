```python
from IPython.display import Markdown
import sympy
from sympy import  Point, Line, Segment, Polygon, RegularPolygon
from sympy import sin as Sin
from sympy import cos as Cos
from sympy import pi as Pi
from sympy import latex as Latex
import matplotlib.pyplot as plt
```

```python
def initAx(ax):
    ax.set_aspect('equal')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position(('data', 0))
    ax.spines['left'].set_position(('data', 0))
    ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
    ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
```

```python
def drawPoint(point, name=None, offset=(0.1, 0.1), color='g'):
    ax = plt.gca()
    initAx(ax)    
    x, y = point.x, point.y
    
    ax.scatter(x, y, color=color)
    
    ax.annotate(
        name,
        xy=(x, y),
        xytext=(x + offset[0], y + offset[1])
    )
```

```python
def drawPolygon(polygon, names=None, offset=(0.1, 0.1), color='g'):
    ax = plt.gca()
    initAx(ax)    
    
    x = [vert.x for vert in polygon.vertices] + [polygon.vertices[0].x]
    y = [vert.y for vert in polygon.vertices] + [polygon.vertices[0].y]
    ax.plot(x, y, color)
    
    for i, vert in enumerate(polygon.vertices):
        drawPoint(vert, names[i] if names != None else "", offset, color)
```

```python
def drawSegment(segment, names=None, offset=(0.1, 0.1), color='g'):
    ax = plt.gca()
    initAx(ax)
    
    points = [segment.p1, segment.p2]
    x = [p.x for p in points]
    y = [p.y for p in points]
    ax.plot(x, y, color)
    
    for i, vert in enumerate(points):
        drawPoint(vert, names[i] if names != None else "", offset, color)
```

```python
def drawIntersction(a, b, names=None):    
    for item in a.intersection(b):
        if type(item) == sympy.Point2D:
            drawPoint(item, color="red")
            if names != None:
                display(Markdown(f"Точка {names[item]}"))
        elif type(item) == sympy.Segment2D:
            drawSegment(item, color="red")
            if names != None:
                display(Markdown(f"Отрезок {names[item.p1]}{names[item.p2]}"))
```

```python
def toRad(deg):
    return deg * Pi/180
```

### Задание 1.
Построить равнобедренный треугольник АВС (АВ = ВС) с вершиной В в начале координат, так что биссектриса ВD образует угол 45 градусов с положительным направлением оси ОХ и 30 градусов со стороной АВ, точка А расположена между осью OY и лучом ВD,  длина АС равна 3. 

У полученного треугольника вывести на экран длины сторон.

Указание: напротив угла 30 градусов в прямоугольном треугольнике лежит катет, равный половине гипотенузы, в равнобедренном треугольнике биссектриса, проведенная к основанию, является медианой и высотой. Координаты точки равны произведению длины отрезка, соединяющего точку с началом координат и косинуса (синуса) угла, образованного этим отрезком и положительным направлением горизонтальной оси.

```python
a = 3 # Т.к. половина угла B равна 30, то весь угол В равен 60 => треугольник равносторонний

B = Point(0, 0)
A = Point(a * Cos(toRad(45 + 30)), a * Sin(toRad(45 + 30)))
C = Point(a * Cos(toRad(45 - 30)), a * Sin(toRad(45 - 30)))
names = {A: "A", B: "B", C: "C"}

ABC = Polygon(A, B, C)

for side in ABC.sides:
    name = names[side.p1] + names[side.p2]
    display(Markdown(f"Длина ${name}$: {side.length.evalf()}"))

names = ["A", "B", "C"]
drawPolygon(ABC, names)
```

### Задание 2.
Построить четырехугольник ABCD, где точки A, B и C из Задания 1, а точка D - центр ABC. Найти площадь, периметр и центр ABCD.
Изобразить ABC и ABCD на графике. Выпуклый ли многоугольник?

```python
A, B, C = ABC.vertices
D = ABC.centroid

ABCD = Polygon(A, B, C, D)

display(Markdown(f"- Площадь $ABCD$: ${Latex(ABCD.area)}$"))
display(Markdown(f"- Периметр $ABCD$: ${Latex(ABCD.perimeter)}$"))
display(Markdown(f"- Центр $ABCD$: $({Latex(ABCD.centroid.x)}, {Latex(ABCD.centroid.y)})$"))
display(Markdown(f"- Выпуклый ли многоугольник? {ABCD.is_convex()}"))

drawPolygon(ABCD, names + ["D"])
drawPolygon(ABC, names, color='blue')
drawPoint(ABCD.centroid, "Center", color="red")
```

### Задание 3
Постоить ромб ALCD с точками A, C и D из Заданий 1 и 2. Найти общую часть треугольника BLC и ромба и вывести на экран в виде имен точек, отрезков и т.п., изобразить на графике  треугольник зеленым, ромб - фиолетовым. 

Указание: диагонали ромба пересекаются в серединах; имя точки искать в словаре имен.

```python
A, B, C = ABC.vertices
D = ABC.centroid
L = D + (Segment(A, C).projection(D) - D).scale(2, 2)
names = {A: "A", B: "B", C: "C", D: "D", L: "L"}

ALCD = Polygon(A, L, C, D)
BLC = Polygon(B, L, C)

drawPolygon(ALCD, ["A", "L", "C", "D"], color = "purple")
drawPolygon(BLC, ["B", "L", "C"])

display(Markdown("Общие части $ALCD$ и $BLC$:"))
drawIntersction(ALCD, BLC, names)
```

### Индивидуальное задание. Вариант 11.

$R = 10$, $\alpha = -90$, $\beta = 45$, O(8, -11).

Построить на одном графике правильный треугольник, шестиугольник и восьмиугольник с радиусом описанной окружности $R$, центром в точке O.

Повернуть треугольник на $\alpha$ градусов, изобразить старый и новый треугольники на одном графике.

Найти общую область правильного шестиугольника и его копии, повернутой на угол $\beta$ градусов.

```python
O = Point(8, -11)
R = 10
alpha = toRad(-90)
beta = toRad(45)

triangle = RegularPolygon(O, R, 3)
hexagon = RegularPolygon(O, R, 6)
octagon = RegularPolygon(O, R, 8)

drawPolygon(triangle)
drawPolygon(hexagon)
drawPolygon(octagon)
plt.show()

newTriangle = triangle.rotate(alpha, O)
drawPolygon(triangle)
drawPolygon(newTriangle, color="orange")
plt.show()

newHexagon = hexagon.rotate(beta, O)
drawPolygon(hexagon)
drawPolygon(newHexagon, color="orange")
drawIntersction(hexagon, newHexagon)
```
