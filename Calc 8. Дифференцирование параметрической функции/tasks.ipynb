```python
from IPython.display import Markdown
import numpy as np
import matplotlib.pyplot as plt
import sympy
from sympy import Symbol, diff, solve, nsolve, solveset, latex, S, simplify, fraction, Union, Interval, intersection, limit
from sympy.calculus.util import continuous_domain
from sympy import sin as Sin
from sympy import cos as Cos
from sympy import pi as Pi
from sympy import exp as Exp
%matplotlib inline

t = Symbol("t")
```

```python
def paramDiff(x, y, order=1):
    dy, dx = diff(y), diff(x)
    
    if order == 1:
        res = dy / dx
    elif order == 2:
        res = (dx*diff(dy) - diff(dx)*dy) / dx**3
    
    return simplify(res)
```

```python
def getExtremumPoints(x, y):
    delta = 1e-5
    mx, mn = [], []
    dy_x = paramDiff(x, y)
    
    for t_0 in list(solveset(dy_x)):
        dy_x_l = dy_x.subs(t, t_0 - delta)
        dy_x_r = dy_x.subs(t, t_0 + delta)
        x_l = x.subs(t, t_0 - delta)
        x_r = x.subs(t, t_0 + delta)

        if (dy_x_l < 0 and dy_x_r > 0 and x_l < x_r) or (dy_x_l > 0 and dy_x_r < 0 and x_l > x_r):
            mn.append(x.subs(t, t_0))

        if (dy_x_l > 0 and dy_x_r < 0 and x_l < x_r) or (dy_x_l < 0 and dy_x_r > 0 and x_l > x_r):
            mx.append(x.subs(t, t_0))
            
    return mx, mn
```

```python
def getInflectionPoints(x, y):
    delta = 1e-5
    points = []
    dy2_x = paramDiff(x, y, 2)
    
    for t_0 in list(solveset(dy2_x)):
        dy2_x_l = dy2_x.subs(t, t_0 - delta)
        dy2_x_r = dy2_x.subs(t, t_0 + delta)

        if dy2_x_l != dy2_x_r:
            points.append(x.subs(t, t_0))
            
    return points
```

$$

тогда производная $y$ по $x$  находится по формуле
$$
y'_x = \frac{y'_t}{x'_t} = \frac{\psi'_t}{\varphi'_t}

$$
y''_{xx} = \frac{x'_t y''_{tt}-x''_{tt}y'_t}{(x'_t)^3}  = \frac{\varphi'_t \psi''_{tt}-\varphi''_{tt}\psi'_t}{(\varphi'_t)^3}
$$

## Асимптоты параметрически заданной функции
Пусть зависимость $y$ от $x$ задана параметрически:
$$
\left\{
\begin{matrix}
x = \varphi(t)\\
y = \psi(t)
\end{matrix}
\right.
$$
Сначала нужно определить, нет ли таких значений параметра $t$, при которых одна из функций $\varphi(t)$ или $\psi(t)$ обращается в бесконечность, а другая остается конечной. При $\varphi(t_0) = \infty$, а $\psi(t_0) = c$ кривая имеет горизонтальную асимптоту  $y = c$. При  $\psi(t_0) = \infty$, а $\varphi(t_0) = c$ кривая имеет вертикальную асимптоту $x = c$. 

Если $\varphi(t_0) = \psi(t_0) = \infty$ и 
$$
\lim_{t\to t_0} \frac{\psi(t)}{\varphi(t)} = k;\qquad \lim_{t\to t_0} [\psi(t) - k\varphi(t)] = b,
$$
то   кривая имеет наклонную асимптоту  $y = kx + b$. 

### Индивидуальное задание. Вариант 11.

Найти асимптоты параметрически заданной функции и изобразить на графике функцию и ее асимптоты.

$$x = 7 e^{t} + 7 e^{- t}$$

$$y = 2 e^{t} - 2 e^{- t}$$

Проанализируем функцию.

При $t \rightarrow \infty$:
- $x \rightarrow \infty$
- $y \rightarrow \infty$

При $t \rightarrow -\infty$:
- $x \rightarrow \infty$
- $y \rightarrow -\infty$

$\Rightarrow$ Скорее всего у функции есть наклонные асимптоты

```python
x = 7*Exp(t) + 7*Exp(-t)
y = 2*Exp(t) - 2*Exp(-t)

k1 = limit(y / x, t, sympy.oo)
b1 = limit(y - k1*x, t, sympy.oo)

k2 = limit(y / x, t, -sympy.oo)
b2 = limit(y - k2*x, t, -sympy.oo)

v = Symbol("v")

asympt1 = k1*v + b1
asympt2 = k2*v + b2
display(asympt1, asympt2)

plt = sympy.plot(show=False)
plt.append(sympy.plot(asympt1, (v, 0, 50), line_color="black", show=False)[0])
plt.append(sympy.plot(asympt2, (v, 0, 50), line_color="black", show=False)[0])
plt.append(sympy.plot_parametric(x, y, (t, -2, 2), show=False)[0])
plt.show()
```
