```python
import sympy
from sympy import S, Symbol, symbols, Function, integrate, diff, Curve, plot, plot_parametric, latex, solveset, Interval, plot_implicit
from sympy import sin as Sin
from sympy import cos as Cos
from sympy import pi as Pi
%matplotlib inline
```

## Занятие 11
## Математический анализ
## Применение интегралов для вычисления площади фигуры и длины дуги 

### Задание 1
Вычислить площадь фигуры, ограниченной графиками функций $f = 2 + x^2$ и  $g = 2 - x^3$. Построить графики, найти точки пересечения, затем вычислить соответствующие интегралы.

```python
x = Symbol('x')
f, g = 2 + x**2, 2 - x**3
a, b = list(solveset(sympy.Eq(f, g)))
plot(f, g, xlim=(a, b), ylim=(0,5))
display(a,b)
abs(integrate(f - g, (x, a, b)))
```

### Задача 2.
Вычислить площадь фигуры, ограниченной графиком функции $x = \cos^3(t)$, $y = \sin^3(t)$, $0 \le t \le 2\pi$.

```python
x, y, t = symbols("x y t")
x, y = Cos(t)**3, Sin(t)**3
plot_parametric(x, y, (t, 0, 2*Pi))
plot_parametric(x, y, (t, 0, Pi))
abs(2*integrate(y*diff(x), (t, 0, Pi)))
```

### Задача 3.
Вычислить площадь фигуры, ограниченной графиком функции $r = \cos(2\phi)$, $-\frac{\pi}{4} \le \phi \le \frac{\pi}{4}$. 

```python
phi = Symbol("phi")
r = Cos(2*phi)
x, y = r*Cos(phi), r*Sin(phi)
plot_parametric(x, y, (phi, -Pi/4, Pi/4))
plot_parametric(x, y, (phi, 0, Pi/4))
abs(2*integrate(y*diff(x), (phi, 0, Pi/4)))
```

### Задача 4.
Вычислить длину кривой, заданной уравнением $y^2 = x^3$ при $x$ от 1 до 3. (Обратите внимание, что у кривой две ветви, симметричные относительно оси OX!)

```python
x, y = symbols("x y")
plot_implicit(sympy.Eq(y**2, x**3))
display(Curve((x, x**(S(3)/2)), (x, 1, 3)).length)
```

### Задача 5.
Вычислить длину кривой, заданной параметрически $x = \frac{3}{2}\cos^3 t $, $y = 3\sin^3 t$.

```python
t = symbols("t")
x, y = S(3)/2 * Cos(t)**3, 3*Sin(t)**3
plot_parametric(x, y)
Curve((x, y), (t, 0, 2*Pi)).length
```

### Задача 6.
Вычислить длину кривой, заданной в полярных координатах
$r = \phi$, $\phi$ от 0 до $2\pi$. Результат вычислить приближенно до 3 знаков после запятой.

```python
phi = Symbol("phi")
r = phi
x, y = r*Sin(phi), r*Cos(phi)
round(Curve((x, y), (phi, 0, 2*Pi)).length.evalf(), 3)
```

### Индивидуальное задание. Вариант 11.

Вычислить площадь фигуры, ограниченной графиками функций $f(x) = 2cos(x), g(x) = 6cos(x)$ на интервале $[0; 2π]$. Построить графики, найти точки пересечения (выделить корни на интервале с помощью intersect и Interval), затем вычислить соответствующие интегралы.

```python
x = symbols("x")
f, g = 2*Cos(x), 6*Cos(x)
plot(f, g, (x, 0, 2*Pi))
a, b = list(solveset(sympy.Eq(f, g), domain=Interval(0, 2*Pi)))
2 * abs(integrate(g - f, (x, a, b)))
```
