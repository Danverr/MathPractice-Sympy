### Задание 1.
Составить в матричном виде двумя способами уравнение поверхности второго порядка $4x^2-24xy+8\sqrt3 xz-7y^2+10\sqrt3 yz+3z^2-4 = 0$. Привести уравнение поверхности второго порядка к каноническому виду с помощью перехода к новому базису. Вывести на экран каноническое уравнение.

```python
x, y, z = symbols("x y z")
expr1 = 4*x**2 - 24*x*y + 8*Sqrt(3)*x*z - 7*y**2 + 10*Sqrt(3)*y*z + 3*z**2 - 4
display(expr1)

A = Matrix([
    [4, -12, 4*Sqrt(3)],
    [-12, -7, 5*Sqrt(3)],
    [4*Sqrt(3), 5*Sqrt(3), 3],
])
b = Matrix([[0, 0, 0]])
c = Matrix([-4])
X = Matrix([x, y, z])

expr2 = simplify((X.T*A*X + 2*b*X + c)[0, 0])
display(A, b, c, expr2)

A = Matrix([
    [4, -12, 4*Sqrt(3), 0],
    [-12, -7, 5*Sqrt(3), 0],
    [4*Sqrt(3), 5*Sqrt(3), 3, 0],
    [0, 0, 0, -4],
])
X = Matrix([x, y, z, 1])

expr3 = simplify((X.T*A*X)[0, 0])
display(A, expr3)

display(expr1 == expr2, expr1 == expr3)
```

```python
A = Matrix([
    [4, -12, 4*Sqrt(3)],
    [-12, -7, 5*Sqrt(3)],
    [4*Sqrt(3), 5*Sqrt(3), 3],
])
c = Matrix([-4])
X = Matrix([x, y, z])

T, J = A.jordan_form()
expr = (X.T*J*X + c)[0, 0]
display(expr)
```

### Задание 2.
Привести уравнение поверхности второго порядка $24 x^{2} - 4 \sqrt{2} x y - 4 \sqrt{2} x z + 18 y^{2} + 12 y z + 18 z^{2} - 4 = 0$ к каноническому виду. Построить график.

```python
A = Matrix([
    [24, -2*Sqrt(2), -2*Sqrt(2)],
    [-2*Sqrt(2), 18, 6],
    [-2*Sqrt(2), 6, 18],
])
c = Matrix([-4])
X = Matrix([x, y, z])

T, J = A.jordan_form()
expr = (X.T*J*X + c)[0, 0]
display(expr)

Z = solve(expr, z)
display(*Z)
plot3d(*Z, (x, -1, 1), (y, -1, 1))
```

### Задание 3.
Привести уравнение поверхности второго порядка $x^2 + y^2 + 2x - 4y + 2z + 1 = 0$ к каноническому виду. Построить график.

```python
A = Matrix([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 0],
])
b = Matrix([1, -2, 1])
c = Matrix([1])
X = Matrix([x, y, z])

T, J = A.jordan_form()
b_new = (T.T*b).T
display(X.T*J*X + 2*b_new*X + c)

for i in range(3):
    if J[i, i] != 0:
        c[0, 0] -= b_new[i]**2 / J[i, i]
        
A = Matrix([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

expr = (X.T*A*X + c)[0]
display(expr)

Z = solve(expr, z)
display(*Z)
plot3d(*Z)
```
