```python
import numpy as np
import sympy
from IPython.display import Markdown
from sympy import S, Symbol, symbols, solve, solveset, plot, plot_parametric, And, plot_implicit
from sympy import sin as Sin
from sympy import cos as Cos
from sympy import exp as Exp
from sympy import log as Log
from sympy import pi as Pi
from sympy.plotting import plot3d
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y, z = symbols("x y z")

def show_countourline_2d(f, x, y):
    plt = plot(ylabel='y', show=False)

    for c, col in [(-2, 'lightgrey'), (-1, 'grey'), (0, 'black'), (1, 'grey'), (2, 'lightgrey')]:
        plt.append(plot_implicit(sympy.Eq(f(x, y), c), line_color=col, line_width=15, show=False)[0])

    plt.show()
    

def show_countourline_3d(f, x, y, z):
    c = symbols('c')
    sol = solve(sympy.Eq(f(x, y, z), c), z)
    
    for const in [-2, 0, 2]:
        plot3d(sol[0].subs(c, const), sol[1].subs(c, const))
```

### Задание 1
Изобразить на рисунках область определения функций

1) $z=\ln(x^2+y)$

2) $z=\sqrt{\sin(x^2+y^2)}$

3) $z=\sqrt{x^2-4}+\sqrt{9-y^2}$

```python
plot_implicit(x**2 + y > 0)
plot_implicit(Sin(x**2 + y**2) >= 0)
plot_implicit(And(x**2 - 4 >= 0, 9 - y**2 >= 0))
```

### Задача 2.
Построить на отдельных графиках линии уровня функций $z=\ln(x^2+y)$ и $z=x^2-y^2$.

```python
def f(x, y):
    return sympy.log(x**2 + y)

def g(x, y):
    return x**2 - y**2


show_countourline_2d(f, x, y)
show_countourline_2d(g, x, y)
```

### Задача 3.
Найти и изобразить поверхности уровня функции $u=(x-1)^2+(y-2)^2+z^2$.

```python
def f(x, y, z):
    return (x - 1)**2 + (y - 2)**2 + z**2

show_countourline_3d(f, x, y, z)
```

### Задача 5.
Найти частные производные по каждому аргументу функции $u=(xy)^z$

```python
expr = (x * y)**z
display(*[expr.diff(t) for t in (x, y, z)])
```

### Задача 6.
Проверить выполнение теоремы Эйлера для функции $f(x,y)=\frac{x}{x^2+y^2} $.
###### Указание.
Упростить выражение с помощью simplify, прежде чем сравнивать.

```python
k = symbols('k', real=True)
expr = x / (x**2 + y**2)

display(Markdown("Однородность: $f(ax, ay) = k^n \cdot f(x, y)$"), sympy.simplify(expr.subs([(x, k*x), (y, k*y)]) / expr))
display(Markdown("Проверим теорему Эйлера"), sympy.simplify(x*expr.diff(x) + y*expr.diff(y)))
```

### Задача 7.
Найти полный дифференциал функции $z=\sin^2x+\cos^2y$

```python
expr = sympy.sin(x)**2 + sympy.cos(y)**2
dx, dy = symbols('dx dy')
expr.diff(x)*dx + expr.diff(y)*dy
```

```python
def f(x, y):
    return -sympy.asin(2*x*y)

show_countourline_2d(f, x, y)

expr = f(x, y)
expr.diff(x)*dx + expr.diff(y)*dy
```
